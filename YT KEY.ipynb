```python
import csv
from google.cloud import vision_v1
from google.cloud.vision_v1 import types
import os
```

```python
# Set the environment variable GOOGLE_APPLICATION_CREDENTIALS to the path of your service account JSON key
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "C:/Users/shrey/OneDrive/Desktop/Thesis/9B/profound-jet-414301-a69c06f48532.json"

def analyze_image(file_path):
    client = vision_v1.ImageAnnotatorClient()

    with open(file_path, "rb") as image_file:
        content = image_file.read()

    image = types.Image(content=content)
    response = client.label_detection(image=image)

    labels = response.label_annotations
    keywords = [label.description.lower() for label in labels]

    return keywords

def process_thumbnails(directory, output_csv):
    with open(output_csv, 'w', newline='') as csvfile:
        fieldnames = ['Thumbnail', 'Keywords']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        for filename in os.listdir(directory):
            if filename.lower().endswith((".png", ".jpg", ".jpeg")):
                file_path = os.path.join(directory, filename)
                keywords = analyze_image(file_path)
                thumbnail_number = filename.split('.')[0]  # Assuming your thumbnail files are named like "1.png", "2.jpg", etc.
                writer.writerow({'Thumbnail': thumbnail_number, 'Keywords': ', '.join(keywords)})
                print(f"Keywords for {filename}: {keywords}")

if __name__ == "__main__":
    thumbnails_directory = "C:/Users/shrey/PycharmProjects/pythonProject6/thumbnailsF4"
    output_csv = "C:/Users/shrey/OneDrive/Desktop/Thesis/9B/result4.csv"
    process_thumbnails(thumbnails_directory, output_csv)
```
