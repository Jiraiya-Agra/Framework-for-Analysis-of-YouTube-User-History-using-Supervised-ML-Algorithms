```python
API_KEY = 'AIzaSyD8CCBTTFdXODTM6zohywGfWt4mT357F8I'  


def get_video_data(api_key, video_url):
    # Check if video_url is not None
    if video_url is not None:
        # Extract the video ID from the URL
        video_id = video_url.split('=')[-1]

        # Sending a request to the YouTube API
        api_url = f'https://www.googleapis.com/youtube/v3/videos?part=snippet,contentDetails,statistics&id={video_id}&key={api_key}'
        response = requests.get(api_url)
        api_data = response.json()

        # Checking if there are any items in the API response
        if 'items' in api_data and len(api_data['items']) > 0:
            # Extract the desired metrics from the response
            try:
                video_title = api_data['items'][0]['snippet']['title']
                video_category_id = api_data['items'][0]['snippet']['categoryId']
                video_category = get_category_name(video_category_id)
                video_description = api_data['items'][0]['snippet']['description']
                video_tags = api_data['items'][0]['snippet'].get('tags', [])
                video_duration = api_data['items'][0]['contentDetails']['duration']
                video_view_count = api_data['items'][0]['statistics'].get('viewCount', 0)
                video_like_count = api_data['items'][0]['statistics'].get('likeCount', 0)
                video_comment_count = api_data['items'][0]['statistics'].get('commentCount', 0)
                return (video_title, video_category, video_category_id, video_description, video_tags,
                        video_duration, video_view_count, video_like_count, video_comment_count)
            except KeyError:
                return None, None, None, None, None, None, None, None, None
        else:
            return None, None, None, None, None, None, None, None, None
    else:
        return None, None, None, None, None, None, None, None, None


def get_category_name(video_category_id):
    # Mapping category IDs to category names (We can extend this dictionary if needed by adding more categories).
    # Currently, there are 32 categories provided by YouTube that are getting used in the category map.
    category_map = {
        '1': 'Film & Animation',
        '2': 'Autos & Vehicles',
        '10': 'Music',
        '15': 'Pets & Animals',
        '18': 'Short Movies',
        '19': 'Travel & Events',
        '17': 'Sports',
        '20': 'Gaming',
        '21': 'Videoblogging',
        '22': 'People & Blogs',
        '23': 'Comedy',
        '24': 'Entertainment',
        '25': 'News & Politics',
        '26': 'Howto & Style',
        '27': 'Education',
        '28': 'Science & Technology',
        '29': 'Nonprofits & Activism',
        '30': 'Movies',
        '31': 'Anime/Animation',
        '33': 'Classics',
        '34': 'Comedy',
        '36': 'Drama',
        '35': 'Documentary',
        '37': 'Family',
        '38': 'Foreign',
        '39': 'Horror',
        '40': 'Sci-Fi/Fantasy',
        '41': 'Thriller',
        '42': 'Shorts',
        '43': 'Shows',
        '44': 'Trailers',

    }
    return category_map.get(video_category_id, 'Unknown Category')


if __name__ == "__main__":
    # Reading video URLs from the JSON file
    with open('C:/Users/shrey/OneDrive/Desktop/Thesis/3B/new_watch_1.JSON', 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Creating an empty list to store video data
    video_data_list = []

    # Processing each video URL
    for item in data:
        if 'titleUrl' in item:
            url = item['titleUrl']
            video_title, video_category, video_category_id, video_description, video_tags, video_duration, video_view_count, video_like_count, video_comment_count = get_video_data(
                API_KEY, url)
            print(f"Video Title: {video_title}")
            print(f"Category: {video_category}")
            print(f"Duration: {video_duration}")
            print(f"View Count: {video_view_count}")
            print(f"Like Count: {video_like_count}")
            print(f"Comment Count: {video_comment_count}")
            print(f"titleUrl: {url}")
            print()

            # Append the data to the list
            video_data_list.append({
                'Video Title': video_title,
                'Category': video_category,
                'Category ID': video_category_id,
                'Description': video_description,
                'Tags': video_tags,
                'Duration': video_duration,
                'View Count': video_view_count,
                'Like Count': video_like_count,
                'Comment Count': video_comment_count,
                'titleUrl': url
            })

    # Creating a DataFrame from the list
    df = pd.DataFrame(video_data_list)

    # Printing the DataFrame
    print(df)
```
